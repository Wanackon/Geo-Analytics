```python
import pandas as pd
import numpy as np

import folium
loc = ' Propensity Score |    1^2'
title_html = '''
             <h1 align="center" style="font-size:16px"><b>{}</b></h3>
             '''.format(loc)
latlon = [[48.48433400021111, 34.92424629064351],[48.47849881938102, 35.01698022827675],[48.46902516836816, 35.04658066084495],[48.46794235276244, 35.04984691547317],[48.4105596042579, 35.00139371050537],[48.40588618028703, 35.03246773645164]]
mapit = folium.Map( location=[48.564884,34.761757], zoom_start=10 )
for coord in latlon:
    folium.Marker( location=[ coord[0], coord[1] ], fill_color='#43d9de', radius=8, popup = coord ).add_to( mapit )
folium.Choropleth(
    geo_data=f,
    data=Geo_agg,
    columns=['Point_ID','Mean_Score'],
    key_on="feature.id",
    fill_color="YlGnBu",
    fill_opacity=0.5,
    line_opacity=0.1,
    legend_name="Point",
    Highlight= True,
    nan_fill_color = "White",
    reset=True,
).add_to(mapit)
mapit.get_root().html.add_child(folium.Element(title_html))
mapit        
```
